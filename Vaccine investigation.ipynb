## 读取各省份发布的的疫苗文件

```python
df1=pd.read_csv('fuckcqcs-master\csv\Jiangxi.csv')
df2=pd.read_csv('fuckcqcs-master\csv\Anhui.csv')
df3=pd.read_csv('fuckcqcs-master\csv\Guangdong.csv')
df4=pd.read_csv('fuckcqcs-master\csv\Xinjiang.csv')
df5=pd.read_csv('fuckcqcs-master\csv\Guizhou.csv')
df6=pd.read_csv('fuckcqcs-master\csv\Hainan.csv')
df7=pd.read_csv('fuckcqcs-master\csv\Heilongjiang.csv')
df8=pd.read_csv('fuckcqcs-master\csv\Jiangsu.csv')
df9=pd.read_csv('fuckcqcs-master\csv\Jilin.csv')
df10=pd.read_csv('fuckcqcs-master\csv\Liaoning.csv')
df11=pd.read_csv('fuckcqcs-master\csv\ENingxia.csv')
df12=pd.read_csv('fuckcqcs-master\csv\Shanxi.csv')
df13=pd.read_csv('fuckcqcs-master\csv\Sichuan.csv')
df14=pd.read_csv('fuckcqcs-master\csv\Tianjing.csv')
df15=pd.read_csv('fuckcqcs-master\csv\Shanghai.csv')
dfzj = pd.read_csv("fuckcqcs-master\csv\Zhejiang.csv")#accomodate a different type csv file to the previous ones
dfzj.drop(columns=['序号','计量单位'],inplace=True)
dfzj.columns=['name', 'create_company', 'price']
dfzj['prov']='浙江省'
df=pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12,df13,df14,df15,dfzj],ignore_index=True)
```

## 分析

### 疫苗一：水痘疫苗 (Chicken-pot)

#### 我们可以看到，长生公司在全国一共生产了25种水痘疫苗（不是指药本身有25种，而是它们年份/使用地区/价格不同），高居第二

#### 图片的乱码问题来不及搞了emmm，大家将就着看一下吧，第二条柱子就是该公司23333

```python
dups = [col for col in df.name if '水痘' in col]#search for vaccine names that contain MMR
for x in dups:
    df.loc[df['name'] == x, 'name'] = '水痘减毒活疫苗1'#assign an identical name to different types of MMR vaccine
df_shuidou = df[df['name']=='水痘减毒活疫苗1']#create a dataframe to exclusively analyze the MMR vaccine
print("生产水痘疫苗的公司以及该公司生产的水痘疫苗类型总数")
print(df_shuidou['create_company'].str.strip().value_counts())
df_shuidou['create_company'].str.strip().value_counts().plot(kind="bar")
```

### 疫苗二：MMR（麻疹风疹腮腺炎联合疫苗）

#### 我们可以看到，该疫苗都是北京的公司所生产，与某公司无直接联系，可放心使用orz

```python
dups = [col for col in df.name if '麻疹' in col or '麻' in col] 
for x in dups:
    df.loc[df['name'] == x, 'name'] = '麻疹风疹联合减毒活疫苗1'#assign an identical name to different types of chicken-pot vaccine
df_mmr = df[df['name']=='麻疹风疹联合减毒活疫苗1']#create a dataframe to exclusively analyze the chicken-pot vaccine
print("生产疫苗的公司以及该公司生产的MMR疫苗类型总数")
print(df_mmr['create_company'].str.strip().value_counts())
```

### 疫苗三：肝炎疫苗（Hepatitis A & B）
#### 又看到了熟悉的名字。。。不过好像量比较少。。。应该不会被我碰上吧。。。

```python
dups = [col for col in df.name if '甲肝' in col or '甲型肝炎' in col or '乙肝' in col or '乙型肝炎' in col or '甲乙' in col]
for x in dups:
    df.loc[df['name'] == x, 'name'] = '肝炎疫苗' #assign an identical name to different types of Hepatitis vaccine
df_he=df[df['name']=='肝炎疫苗'] #create a dataframe to exclusively analyze the Hepatitis vaccine
print("生产疫苗的公司以及该公司生产的肝炎疫苗类型总数")
print(df_he['create_company'].str.strip().value_counts().head(10))
```
